# Agglomerative Clustering
Main objective to categorize similar datapoints togethor

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1: Read the dataset

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

### Problem statement - Segment the similar customers on basis of Age, Income and Spending

### Step 2: Perform Basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3: Seperate X 

```python
X = df[['Age', 'Income', 'Spending']]
X.head()
```

### Step 4: Create a Preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Selecting Optimum number of clusters using silhoutee scores

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
n = list(range(2, 15))
n
```

```python
sil = []
for i in n:
    # Create a model
    model = AgglomerativeClustering(n_clusters=i)
    # Predict the labels
    lab = model.fit_predict(X_pre)
    # Calculate silhoutee score
    s = silhouette_score(X_pre, lab)
    # Append into sil list
    sil.append(s)
```

```python
sil
```

```python
n
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(n, sil)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutee Score')
plt.title('Silhoutee score for different clusters')
plt.grid()
plt.show()
```

### From above graph highest silhoutee score of 0.42 is selected with n_clusters=6

### Create a final model with n clusters=6

```python
final_model = AgglomerativeClustering(n_clusters=6)
clusters = final_model.fit_predict(X_pre)
clusters
```

```python
X.head()
```

```python
X['Label'] = clusters
```

```python
X.head()
```

```python
for i in range(6):
    print(f'Cluster Number {i}: \n')
    display(X[X['Label']==i])
    print('\n===================================================\n')
```

```python
centroids = X.groupby(by='Label').mean()
```
