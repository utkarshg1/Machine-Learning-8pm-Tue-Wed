# Time Series Decomposition


```python
from warnings import filterwarnings
filterwarnings('ignore')
```

# Read the dataset

```python
import pandas as pd
df = pd.read_csv('AirPassengers.csv', index_col='Month', parse_dates=True)
df.head()
```

```python
df.index
```

```python
df.info()
```

### Visualize above data

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(10, 5))
plt.plot(df['Passengers'])
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title('Air Passengers Time Series Data')
plt.grid()
plt.show()
```

### Time Series Decompositon

```python
from statsmodels.tsa.seasonal import seasonal_decompose
Q = seasonal_decompose(df['Passengers'], model='multiplicative')
Q
```

```python
Q.plot()
```

### Trendline 

```python
plt.figure(figsize=(10, 5))
Q.observed.plot()
Q.trend.plot()
plt.xlabel('Month')
plt.ylabel('Number of Air Passengers')
plt.title('Trend analysis')
plt.grid()
plt.show()
```

### Seasonal Plot

```python
plt.figure(figsize=(10, 5))
Q.seasonal.plot()
plt.xlabel('Month')
plt.ylabel('Number of Air Passengers')
plt.title('Seasonal analysis')
plt.grid()
plt.show()
```

### Residual Plot

```python
plt.figure(figsize=(10, 5))
Q.resid.plot()
plt.xlabel('Month')
plt.ylabel('Number of Air Passengers')
plt.title('Residual analysis')
plt.grid()
plt.show()
```

### Time series decompositon is complete

### What is future demand for next 3 years - Forecasting

```python
%pip install pmdarima
```

### Restart kernel and run all cells

```python
from pmdarima.arima import auto_arima
model = auto_arima(df['Passengers'], m=12, seasonal=True)
```

```python
model.order
```

```python
forecast, conf = model.predict(n_periods=48, return_conf_int=True)
```

```python
forecast
```

```python
conf
```

```python
df
```

```python
df_forecast = forecast.to_frame()
df_forecast.columns = ['Forecast']
df_forecast
```

```python
df_forecast[['Lower', 'Upper']] = conf
```

```python
df_forecast
```

### Number of passengers is a whole number so round off above values to nearest integer

```python
df_forecast = df_forecast.apply(round)
```

```python
df_forecast
```

### Save above to csv

```python
df_forecast.to_csv('Forecast.csv')
```

### Plotting above forecast results

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Passengers'], label='Actual')
plt.plot(df_forecast['Forecast'], label='Forecast')
plt.fill_between(df_forecast.index, df_forecast['Lower'], df_forecast['Upper'],
                 color='gray', alpha=0.7, label='Confidence Interval')
plt.xlabel('Month')
plt.ylabel('Number of Air Passengers')
plt.title('Forecast Sheet')
plt.grid()
plt.legend(loc='upper center')
plt.show()
```
